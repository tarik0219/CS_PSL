```python
import matplotlib.pyplot as plt
import numpy as np
import random
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.neighbors import KNeighborsClassifier
import warnings
import matplotlib.pyplot as plt
from pandas import DataFrame
import seaborn as sns
warnings.filterwarnings("ignore")
```

# Creating Data

```python
def create_data(m_random_low,m_random_high,s,csize,size):
    mean_x1 = {}
    mean_x2 = {}
    m1 = random.randint(m_random_low,m_random_high)
    m2 = random.randint(m_random_low,m_random_high)
    mean_x1[0] = m1
    mean_x2[0] = m2
    total1 = np.random.normal(m1,s,size)
    total2 = np.random.normal(m2,s,size)
    for i in range(1,csize):
        m1 = random.randint(m_random_low,m_random_high)
        m2 = random.randint(m_random_low,m_random_high)
        mean_x1[i] = m1
        mean_x2[i] = m2
        test1 = np.random.normal(m1,s,size)
        test2 = np.random.normal(m2,s,size)
        

        total1 = np.concatenate((total1, test1), axis=0)
        total2 = np.concatenate((total2, test2), axis=0)
    return total1,total2, mean_x1, mean_x2
```

```python
x1, x2, mean_0_x1, mean_0_x2 = create_data(20,100,8,10,510)
y1, y2, mean_1_x1, mean_1_x2 = create_data(0,80,8,10,510)
fig, axes = plt.subplots(1, 1)
plt.plot(x1,x2,'ro')
plt.plot(y1,y2,'bo')
fig.set_size_inches(30, 30)
plt.savefig('fig1.jpg')
plt.show()
data_X = np.zeros((10200,2))
data_y = np.zeros((10200,1))
data_X[0:5100,0] = x1
data_X[0:5100,1] = x2
data_X[5100:,0] = y1
data_X[5100:,1] = y2
data_y[0:5100] = 0
data_y[5100:] = 1
X_train, X_test, y_train, y_test = train_test_split(data_X, data_y, test_size=(200/10200), random_state=42)
```

# kNN Best K

```python
print(data_X.shape)
data = np.zeros(shape = (10200,3))
data[:,0:2] = data_X
data[:,2:] = data_y
np.random.shuffle(data)
data
```

```python
num = int(10200/10)
k_dict = {}
for k in range(1,20):
    total_score = 0
    for i in range(1,11):
        X_test_knn = np.zeros(shape = (num,2))
        y_test_knn = np.zeros(shape = (num,1))
        X_train_knn = np.zeros(shape = (num * 9,2))
        y_train_knn = np.zeros(shape = (num * 9,1))
        if i == 1:
            X_test_knn = data[0:num,0:2]
            y_test_knn = data[0:num,2:]
            X_train_knn = data[num:,0:2]
            y_train_knn = data[num:,2:]
            neigh = KNeighborsClassifier(n_neighbors= k)
            neigh.fit(X_train_knn, y_train_knn)
            score = neigh.score(X_test_knn, y_test_knn, sample_weight=None)
            total_score += score
            #print(score)

        else:
            X_test_knn = data[num * (i-1):num * i,0:2]
            y_test_knn = data[num * (i-1):num * i,2:]
            X_train_knn[:num * (i-1), 0:2] = data[:num * (i-1), 0:2]
            X_train_knn[num * (i-1):, 0:2] = data[num * i:, 0:2]
            y_train_knn[:num * (i-1),:] = data[:num * (i-1), 2:]
            y_train_knn[num * (i-1):,:] = data[num * i:, 2:]
            neigh = KNeighborsClassifier(n_neighbors= k)
            neigh.fit(X_train_knn, y_train_knn)
            score = neigh.score(X_test_knn, y_test_knn, sample_weight=None)
            total_score += score
            #print(score)
    k_dict[k] = 1- (total_score/10)
```

```python
x = k_dict.keys()
y = k_dict.values()
plt.scatter(x,y)
plt.show()
```

# Linear Regression

```python
def train_linear(X_train,y_train):
    regr = linear_model.LinearRegression()
    regr.fit(X_train, y_train)
    return regr

def test_linear(regr,X_test,y_test):
    error = 0
    for i,test in  enumerate(X_test):
        predict = regr.predict([test])[0][0]
        if predict < .5:
            predict_score = 0.0
        else:
            predict_score = 1.0
        real_score = y_test[i][0]

        if real_score != predict_score:
            error += 1
    return (error/y_test.shape[0])
```

```python
regr = train_linear(X_train,y_train)
test_linear(regr,X_test,y_test)
```

# Quadratic Regression


```python
def train_quadratic(X_train,y_train):
    poly_reg = PolynomialFeatures(degree = 2) 
    X_poly = poly_reg.fit_transform(X_train)
    poly_reg.fit(X_poly, y_train)
    lin_reg_2 = linear_model.LinearRegression()
    lin_reg_2.fit(X_poly, y_train)
    return lin_reg_2

def test_quadratic(lin_reg_2,X_test,y_test):
    error = 0
    for i,test in  enumerate(X_test):
        predict = lin_reg_2.predict(poly_reg.fit_transform([test]))[0][0]#regr.predict([test])[0][0]
        if predict < .5:
            predict_score = 0.0
        else:
            predict_score = 1.0
        real_score = y_test[i][0]

        if real_score != predict_score:
            error += 1
    return (error/y_test.shape[0])
```

```python
regr_q = train_quadratic(X_train,y_train)
test_quadratic(regr_q,X_test,y_test)
```

# Bayes Error

```python
def test_bayes_error(X_test,y_test, mean_1_x1, mean_1_x2,mean_0_x1,mean_0_x2):
    total_error = 0
    s = 10
    for i,X_value in enumerate(X_test):
        y_value = y_test[i]

        num = 0
        den = 0
        for l in range(0,10):
            m_1 = np.zeros(shape = (1,2))
            m_1[0,0] = mean_1_x1[l]
            m_1[0,1] = mean_1_x2[l]

            m_2 = np.zeros(shape = (1,2))
            m_2[0,0] = mean_0_x1[l]
            m_2[0,1] = mean_0_x2[l]

            num += np.exp((-1/(2*s**2)  * np.linalg.norm(m_1 - X_value) )) 
            den += np.exp((-1/(2*s**2)  * np.linalg.norm(m_2 - X_value) )) 

        #print(num/den, y_value[0])
        if num > den:
            predict = 1.0
        else:
            predict = 0.0

        if y_value[0] != predict:
            total_error += 1

    return total_error/y_test.shape[0]
```

```python
test_bayes_error(X_test,y_test, mean_1_x1, mean_1_x2,mean_0_x1,mean_0_x2)
```

# kNN

```python
def train_knn(X_train,y_train,k):
    neigh = KNeighborsClassifier(n_neighbors= k)
    neigh.fit(X_train, y_train)
    return neigh
def test_knn(neigh,X_test,y_test):
    score = neigh.score(X_test, y_test, sample_weight=None)
    return 1 - score
```

```python
neigh = train_knn(X_train,y_train,14)
test_knn(neigh,X_test,y_test)
```

```python
def simulate(x,k):
    lin_train = []
    lin_test = []
    quad_train = []
    quad_test = []
    knn_train = []
    knn_test = []
    bayes_train = []
    bayes_test = []
    for i in range(0,x):
        #create data
        x1, x2, mean_0_x1, mean_0_x2 = create_data(20,100,8,10,510)
        y1, y2, mean_1_x1, mean_1_x2 = create_data(0,80,8,10,510)
        data_X = np.zeros((10200,2))
        data_y = np.zeros((10200,1))
        data_X[0:5100,0] = x1
        data_X[0:5100,1] = x2
        data_X[5100:,0] = y1
        data_X[5100:,1] = y2
        data_y[0:5100] = 0
        data_y[5100:] = 1
        X_train, X_test, y_train, y_test = train_test_split(data_X, data_y, test_size=(200/10200), random_state=42)
        
        #train each 
        regr = train_linear(X_train,y_train)
        regr_q = train_quadratic(X_train,y_train)
        neigh = train_knn(X_train,y_train,16)
        
        #test test
        lin_test.append(test_linear(regr,X_test,y_test))
        quad_test.append(test_quadratic(regr_q,X_test,y_test))
        knn_test.append(test_knn(neigh,X_test,y_test))
        bayes_test.append(test_bayes_error(X_test,y_test, mean_1_x1, mean_1_x2,mean_0_x1,mean_0_x2))
        
        #test train
        lin_train.append(test_linear(regr,X_train,y_train))
        quad_train.append(test_quadratic(regr_q,X_train,y_train))
        knn_train.append(test_knn(neigh,X_train,y_train))
        bayes_train.append(test_bayes_error(X_train,y_train, mean_1_x1, mean_1_x2,mean_0_x1,mean_0_x2))
        
        print(i)
    
    
    return lin_train,lin_test,quad_train,quad_test,knn_train,knn_test,bayes_train,bayes_test
```

```python
lin_train,lin_test,quad_train,quad_test,knn_train,knn_test,bayes_train,bayes_test = simulate(20,14)
```
